```python
#SSD,HDD,HYBRID,FLASH STORAGE
df['Memory']=df['Memory'].str.replace('GB','')
df['Memory']=df['Memory'].str.replace('TB','000')
df['Memory']=df['Memory'].str.replace(r'\.0','',regex=True)


new=df['Memory'].str.split('+',n=1,expand=True)
df['first']=new[0]
df['first']=df['first'].str.strip()
df['second']=new[1]
df['second']=df['second'].fillna("0")

df['first_SSD']=df['first'].apply(lambda x: 1 if 'SSD' in x else 0)
df['first_HDD']=df['first'].apply(lambda x: 1 if 'HDD' in x else 0)
df['first_Flash_Storage']=df['first'].apply(lambda x: 1 if 'Flash Storage' in x else 0)
df['first_Hybrid']=df['first'].apply(lambda x: 1 if 'Hybrid' in x else 0)

df['second_SSD']=df['second'].apply(lambda x: 1 if 'SSD' in x else 0)
df['second_HDD']=df['second'].apply(lambda x: 1 if 'HDD' in x else 0)
df['second_Flash_Storage']=df['second'].apply(lambda x: 1 if 'Flash Storage' in x else 0)
df['second_Hybrid']=df['second'].apply(lambda x: 1 if 'Hybrid' in x else 0)

df['first']=df['first'].str.replace(r'\D','',regex=True)
df['first']=df['first'].astype(int)

df['second']=df['second'].str.replace(r'\D','',regex=True)
df['second']=df['second'].astype(int)

df['HDD']=(df['first']*df['first_HDD'])+(df['second']*df['second_HDD'])
df['SSD']=(df['first']*df['first_SSD'])+(df['second']*df['second_SSD'])
df['Hybrid']=(df['first']*df['first_Hybrid'])+(df['second']*df['second_Hybrid'])
df['Flash_Storage']=(df['first']*df['first_Flash_Storage'])+(df['second']*df['second_Flash_Storage'])



```

```python
X=df.drop(columns=['Price'])
y=np.log(df['Price'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.15,random_state=2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score,mean_absolute_error
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,ExtraTreesRegressor,VotingRegressor,StackingRegressor,GradientBoostingRegressor,AdaBoostRegressor
from xgboost import XGBRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
```

```python
df.head()
```

```python
ohe=ColumnTransformer(transformers=[('ohe_tnf',OneHotEncoder(sparse_output=False,drop='first'),[0,1,10,14,15])],
                     remainder='passthrough')

linear_regression=LinearRegression()
ridge=Ridge(alpha=10)
lasso=Lasso(alpha=0.001)
knn=KNeighborsRegressor(n_neighbors=3)
dtr=DecisionTreeRegressor(max_depth=8)
svr=SVR(kernel='rbf',C=10000,epsilon=0.1)
random_forest_reg=RandomForestRegressor(n_estimators=100,
                         max_samples=0.5,
                         max_features=0.75,
                         random_state=3,
                         max_depth=15,
                         bootstrap=True)
extra_trees_reg=ExtraTreesRegressor(n_estimators=100,
                         max_samples=0.5,
                         max_features=0.75,
                         random_state=3,
                         max_depth=15,
                        bootstrap=True)
adaboost_reg=AdaBoostRegressor(n_estimators=15,
                              learning_rate=1.0)
gbr=GradientBoostingRegressor(n_estimators=500)
xgbr=XGBRegressor(n_estimators=45,max_depth=5,learning_rate=0.5)
voting_reg=VotingRegressor([('random_forest_reg',random_forest_reg),('extra_trees_reg',extra_trees_reg),
                           ('gbr',gbr),('xgbr',xgbr)],weights=[5,1,1,1])
stacking_reg=StackingRegressor(estimators=[('random_forest_reg',random_forest_reg),
                                          ('gbr',gbr),('xgbr',xgbr)],final_estimator=Ridge(alpha=100))
```

```python
models_list={'linear_regression':linear_regression,
            'ridge':ridge,
            'lasso':lasso,
            'knn':knn,
            'dtr':dtr,
            'svr':svr,
            'random_forest_reg':random_forest_reg,
            'extra_trees_reg':extra_trees_reg,
            'adaboost_reg':adaboost_reg,
            'gbr':gbr,
            'xgbr':xgbr,
            'voting_reg':voting_reg,
            'stacking_reg':stacking_reg}

```

```python
def train_model(model,x_train,x_test,y_train,y_test):
    pipe=Pipeline([
    ('ohe',ohe),
    ('model',model)])
    pipe.fit(x_train,y_train)
    y_pred=pipe.predict(x_test)
    r_squared_score=r2_score(y_test,y_pred)
    mae=mean_absolute_error(y_test,y_pred)
    
    return np.round(r_squared_score,2),np.round(mae,2)
    
```

```python
r_squared_score_list=[]
mae_list=[]

for name,value in models_list.items():
    r_squared_score,mae=train_model(value,x_train,x_test,y_train,y_test)
    
    print('For ',name ,' : ')
    print('r2 Score: ',r_squared_score)
    print('MAE: ',mae)
    
    r_squared_score_list.append(r_squared_score)
    mae_list.append(mae)

    
```

```python
performance_df=pd.DataFrame({'Algorithm':models_list.keys(),
                            'r2 Score':r_squared_score_list,
                            'MAE':mae_list}).sort_values('r2 Score',ascending=False)
```
